```python
import pandas as pd
import numpy as np
import copy
from sklearn.preprocessing import minmax_scale
```

```python
bilstm_bag = read_scale('output/sp_bilstm_relu_d1_atten_global_amsgrad_bag_rmnum_ft125.csv')
omb985 = read_scale('output/one_more_blend_985.csv')
cnn_bag = read_scale('output/sp_cnn_f256_k2345_d1_400_50_relu_global_bag_rmnum_ft225.csv')
bigru = read_scale('output/sp_bigru_ft100atten_all.csv')
bigru_bag = read_scale('output/sp_bigru_relu_ft100_atten_global_nadam_bag_rmnum_sw.csv')
lgb = read_scale('output/lgb.csv')
cnn300_bag = read_scale('output/sp_cnn_f300_k2345_d1_250_50_rs_global_bag_rmnum_ft300.csv')
cnn = read_scale('output/sp_cnn_f300_k2345_d1_250_50_rs_global_all_rmnum_ft300.csv')
sub = copy.deepcopy(bigru_bag)
```

```python
def read_scale(file_path):
    
    df = pd.read_csv(file_path)
    
    col = df.columns[1:]
    content= minmax_scale(df[col].values)
    df[col] = content
    return df
```

```python
col = sub.columns[1:]
sub[col]= (bilstm_bag[col]  +   omb985[col]*3 + cnn_bag[col]*2 + bigru[col]+ bigru_bag[col] + lgb[col] +cnn300[col] ) /10
sub.head()
```

```python
sub.to_csv('output/bilstm_bag_3omb_2cnn_bag_bigru_bigru_bag_lgb_cnn300_bag_scale.csv', index=None)
```
