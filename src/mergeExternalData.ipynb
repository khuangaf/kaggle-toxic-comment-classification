```python
ets = ['insult.tsv', 'threat.tsv','toxicity.tsv']
```

```python
df = pd.read_csv('train.csv')
print df.shape
columns = df.columns
for et in ets:
    etDf = pd.DataFrame(columns = columns)
    etDf.comment_text = pd.read_csv(et, sep='\t').comment
    etDf.loc[:,columns[2:]] = 0
    if 'insult' in et:
        etDf.insult = 1
    elif 'threat' in et:
        etDf.threat = 1
    else:
        etDf.toxic = 1
    print etDf.shape
    df= pd.concat([df,etDf])
print df.shape
df.head()
```

```python
df.to_csv('train_external.csv',index=None)
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
train = pd.read_csv('train.csv')
train.head()
stemmer = SnowballStemmer("english")
train.comment_text = train.comment_text.apply(lambda x: stemmer.stem(x.decode('utf-8')))
train.head()
```

```python
train.to_csv('train_snowball.csv',index=None , encoding='utf-8')
```
